```python
new_reviews = pd.read_csv('books.csv')
new_reviews = new_reviews.dropna(how='any')
new_reviews.head()
```

```python
merged = new_reviews.merge(nyt, on='isbn', how='inner')
merged
```

```python
books = []

for index, row in merged.iterrows():
    title = row['original_title']
    books.append(title)
    
```

```python
base_url = "http://www.omdbapi.com/?" 

query_url = f"{base_url}apikey={omdb_api}&t={books[0]}"

response = requests.get(query_url)

pprint(response.json())
```

```python
json_list = []

for book in books:
    try:
        query_url= base_url + "apikey=" + "trilogy" + "&t=" + book
        response = requests.get(query_url)
        data = response.json()
        json_list.append(data)
        
    except:
        print("Error: Book not made into movie yet...")
        
```

```python
pprint(json_list)
```

```python
df = pd.DataFrame(json_list)
df.head()
```

```python
df = df.dropna(thresh=3)
```

```python
ratings = df[['Title','imdbRating']]
ratings = ratings.reset_index()
```
