```python
#import dependencies
import pandas as pd 
import numpy as np
import requests
from config import omdb_api
from pprint import pprint

pd.set_option('display.float_format', lambda x: '%.0f' % x)
```

```python
#read csv 
goodbooks = pd.read_csv("books.csv")
goodbooks = goodbooks.dropna(how='any')
goodbooks
```

```python
#add titles into a list and avoid duplicates 
titles = []

for index, row in goodbooks.iterrows():
    title = row['original_title']
    if title not in titles:
        titles.append(title)
```

```python
len(titles)
```

```python
titles_1 = titles[0:1000]
```

```python
base_url = "http://www.omdbapi.com/?"
query_url= base_url + "apikey=" + "trilogy" + "&t=" + titles_1[0]
response = requests.get(query_url)
data = response.json()
pprint(data)
```

```python
json_list = []

base_url = "http://www.omdbapi.com/?"

for title in titles:
    try:
        query_url= base_url + "apikey=" + "trilogy" + "&t=" + title
        response = requests.get(query_url)
        data = response.json()
        json_list.append(data)
    except:
        print("Error")
```

```python
#add the list into a dataframe 
df = pd.DataFrame(json_list)
```

```python
df = df.dropna(thresh=3)
```

```python
#slice dataframe for just the required columns
ratings = df[['Title','imdbRating']]
ratings = ratings.reset_index()
```

```python
#replace N/A with NaN to drop those rows 
ratings = ratings.replace('N/A', np.nan)

#drop NaN values
ratings = ratings.dropna(how='any')
```

```python
ratings = ratings[['Title','imdbRating']]
```

```python
ratings.to_csv("output/movie_ratings_1000.csv")
```
