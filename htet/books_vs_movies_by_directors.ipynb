```python
import pandas as pd 
import numpy as np
import requests
# from config import omdb_api
from pprint import pprint

pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
books = pd.read_csv("../../br.csv/br.csv", encoding="utf-8", engine="python", error_bad_lines=False)
```

```python
more_reviews = books.drop(books[books.ratingsCount < 100000].index)
```

```python
more_reviews['title'].nunique()
```

```python
sample = more_reviews.iloc[0:2000]
sample = sample.reset_index()
book_rating = sample[['rating']]
```

```python
books = []

for index, row in sample.iterrows():
    title = row['title']
    books.append(title)
    
```

```python
base_url = "http://www.omdbapi.com/?" 

query_url = f"{base_url}apikey=trilogy&t={books[0]}"

response = requests.get(query_url)

pprint(response.json())
```

```python
json_list = []

for book in books:
    try:
        query_url= base_url + "apikey=" + "trilogy" + "&t=" + book
        response = requests.get(query_url)
        data = response.json()
        json_list.append(data)
        
    except:
        print("...")
```

```python
# pprint(json_list)
```

```python
df = pd.DataFrame(json_list)
df['Book Rating'] = book_rating
df
```

```python
df.columns
```

```python
by_dir= df[['Title','Director','Book Rating', 'imdbRating']]
by_dir['Book Rating'] = by_dir['Book Rating']*2
```

```python
by_dir = by_dir.replace('N/A', np.nan)
by_dir = by_dir.dropna()
```

```python
by_dir['imdbRating'] = by_dir['imdbRating'].astype(float)
```

```python
by_dir = by_dir.reset_index()
```

```python
by_dir = by_dir.drop(columns=['index'])
```

```python
final_df = by_dir.groupby('Director').mean()
```
